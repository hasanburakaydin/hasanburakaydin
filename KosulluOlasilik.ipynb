**KoÅŸullu OlasÄ±lÄ±k (Bayes Teoremi) Ã–rnek Soru**

KumaÅŸ Ã¼reten bir iÅŸletmede toplam Ã¼retimin %35â€™ini ğ‘€1, %25â€™ni ğ‘€2 ve %40â€™nÄ± ğ‘€3 makinesi Ã¼retmektedir. Bu makinelerin Ã¼rettiÄŸi bir top kumaÅŸÄ±n kusurlu olma olasÄ±lÄ±klarÄ± sÄ±rasÄ± ile %1 ,%3 ve %2â€™dir. 

Bir gÃ¼nlÃ¼k Ã¼retimin sonunda bir top kumaÅŸ seÃ§iliyor ve kusurlu olup olmadÄ±ÄŸÄ±na bakÄ±lÄ±yor. KumaÅŸ kusurlu ise; ğ‘€1 makinesinde Ã¼retilmiÅŸ olma olasÄ±lÄ±ÄŸÄ± nedir?


```python
# KoÅŸullu OlasÄ±lÄ±k (Bayes Teoremi) Ã‡alÄ±ÅŸmasÄ± 205112015 Hasan Burak AYDIN

import time

'''
M1 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M1KumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M1KumasYuzde = float(input("M1 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M1KumasYuzde = M1KumasYuzde / 100
print("M1 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M1KumasYuzde,2))+"\n")

'''
M2 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M2KumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M2KumasYuzde = float(input("M2 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M2KumasYuzde = M2KumasYuzde / 100
print("M2 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M2KumasYuzde,2))+"\n")

'''
M3 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M3KumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M3KumasYuzde = float(input("M3 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M3KumasYuzde = M3KumasYuzde / 100
print("M3 Makinesinin KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M3KumasYuzde,2))+"\n")

'''
M1 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M1KusurluKumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M1KusurluKumasYuzde = float(input("M1 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M1KusurluKumasYuzde = M1KusurluKumasYuzde / 100
print("M1 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M1KusurluKumasYuzde,2))+"\n")

'''
M2 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M2KusurluKumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M2KusurluKumasYuzde = float(input("M2 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M2KusurluKumasYuzde = M2KusurluKumasYuzde / 100
print("M2 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M2KusurluKumasYuzde,2))+"\n")

'''
M3 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesinin KullanÄ±cdan AlÄ±nÄ±p;
M3KusurluKumasYuzde DeÄŸiÅŸkenine AtanmasÄ±
'''

M3KusurluKumasYuzde = float(input("M3 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesini Giriniz:"))
M3KusurluKumasYuzde = M3KusurluKumasYuzde / 100
print("M3 Makinesinin Kusurlu KumaÅŸ Ãœretim YÃ¼zdesi: "+str(round(M3KusurluKumasYuzde,2))+"\n")

print("Ãœretilen KumaÅŸ Kusurlu Ä°se KumaÅŸÄ±n M1 Makinesinde Ãœretilme OlasÄ±lÄ±ÄŸÄ± Nedir?\n\n")


time.sleep(3)

'''
1.AdÄ±m Ãœretilen KumaÅŸÄ±n Kusurlu Olma OlaslÄ±ÄŸÄ±:
Bu Durum Ä°stenilen KoÅŸullu OlasÄ±lÄ±ÄŸÄ±n Ã–rnek Uzay (Sample Space)\'ini OluÅŸturur.
'''

'''
Ãœretilen KumaÅŸÄ±n Kusurlu Olma OlasÄ±lÄ±ÄŸÄ±nÄ±n HesaplanÄ±p;
KusurluKumas DeÄŸiÅŸkenine AtanmasÄ±
'''

KusurluKumas = M1KumasYuzde*M1KusurluKumasYuzde + M2KumasYuzde*M2KusurluKumasYuzde + M3KumasYuzde*M3KusurluKumasYuzde
KusurluKumasMetin = "Ãœretilen KumaÅŸÄ±n Kusurlu Olma OlasÄ±lÄ±ÄŸÄ±: {:.2}\n".format(KusurluKumas)
print(KusurluKumasMetin)

'''
2.AdÄ±m Ãœretilen KumaÅŸÄ±n M1 Makinesinde Ãœretilip Kusurlu Olma OlaslÄ±ÄŸÄ±:
Bu Durum Ä°stenilen KoÅŸullu OlasÄ±lÄ±ÄŸÄ±n Olay UzayÄ± (Event Space)\'ini OluÅŸturur.
'''

'''
Ãœretilen KumaÅŸÄ±n M1 Makinesinde Ãœretilip Kusurlu Olma OlasÄ±lÄ±ÄŸÄ±nÄ±n HesaplanÄ±p;
M1KusurluKumas DeÄŸiÅŸkenine AtanmasÄ±
'''

M1KusurluKumas = M1KumasYuzde*M1KusurluKumasYuzde
M1KusurluKumasMetin = "Ãœretilen KumaÅŸÄ±n M1 Makinesinde Ãœretilip Kusurlu Olma OlasÄ±lÄ±ÄŸÄ±: {:.4}\n".format(M1KusurluKumas)
print(M1KusurluKumasMetin)

'''
3. AdÄ±m Bir OlayÄ±n OlasÄ±lÄ±ÄŸÄ±:
Olay UzayÄ± (Event Space)\'in Ã–rnek Uzay (Sample Space)\'e OranlanmasÄ± OlduÄŸu Ä°Ã§in
'''

'''
Ãœretilen KumaÅŸÄ±n Kusurlu OlduÄŸu BilndiÄŸine GÃ¶re M1 Makinesinde ÃœretilmiÅŸ Olma OlasÄ±lÄ±ÄŸÄ±nÄ±n HesaplanÄ±p;
KosulluKusurluKumas DeÄŸiÅŸkenine AtanmasÄ±
'''

KosulluKusurluKumas = M1KusurluKumas / KusurluKumas
KosulluKusurluKumasMetin = "Ãœretilen KumaÅŸÄ±n Kusurlu OlduÄŸu BilndiÄŸine GÃ¶re M1 Makinesinde ÃœretilmiÅŸ Olma OlasÄ±lÄ±ÄŸÄ±: {:.4}".format(KosulluKusurluKumas)
print(KosulluKusurluKumasMetin)
```
