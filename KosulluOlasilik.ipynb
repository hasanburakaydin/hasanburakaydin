**Koşullu Olasılık (Bayes Teoremi) Örnek Soru**

Kumaş üreten bir işletmede toplam üretimin %35’ini 𝑀1, %25’ni 𝑀2 ve %40’nı 𝑀3 makinesi üretmektedir. Bu makinelerin ürettiği bir top kumaşın kusurlu olma olasılıkları sırası ile %1 ,%3 ve %2’dir. 

Bir günlük üretimin sonunda bir top kumaş seçiliyor ve kusurlu olup olmadığına bakılıyor. Kumaş kusurlu ise; 𝑀1 makinesinde üretilmiş olma olasılığı nedir?


```python
# Koşullu Olasılık (Bayes Teoremi) Çalışması 205112015 Hasan Burak AYDIN

import time

'''
M1 Makinesinin Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M1KumasYuzde Değişkenine Atanması
'''

M1KumasYuzde = float(input("M1 Makinesinin Kumaş Üretim Yüzdesini Giriniz:"))
M1KumasYuzde = M1KumasYuzde / 100
print("M1 Makinesinin Kumaş Üretim Yüzdesi: "+str(round(M1KumasYuzde,2))+"\n")

'''
M2 Makinesinin Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M2KumasYuzde Değişkenine Atanması
'''

M2KumasYuzde = float(input("M2 Makinesinin Kumaş Üretim Yüzdesini Giriniz:"))
M2KumasYuzde = M2KumasYuzde / 100
print("M2 Makinesinin Kumaş Üretim Yüzdesi: "+str(round(M2KumasYuzde,2))+"\n")

'''
M3 Makinesinin Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M3KumasYuzde Değişkenine Atanması
'''

M3KumasYuzde = float(input("M3 Makinesinin Kumaş Üretim Yüzdesini Giriniz:"))
M3KumasYuzde = M3KumasYuzde / 100
print("M3 Makinesinin Kumaş Üretim Yüzdesi: "+str(round(M3KumasYuzde,2))+"\n")

'''
M1 Makinesinin Kusurlu Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M1KusurluKumasYuzde Değişkenine Atanması
'''

M1KusurluKumasYuzde = float(input("M1 Makinesinin Kusurlu Kumaş Üretim Yüzdesini Giriniz:"))
M1KusurluKumasYuzde = M1KusurluKumasYuzde / 100
print("M1 Makinesinin Kusurlu Kumaş Üretim Yüzdesi: "+str(round(M1KusurluKumasYuzde,2))+"\n")

'''
M2 Makinesinin Kusurlu Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M2KusurluKumasYuzde Değişkenine Atanması
'''

M2KusurluKumasYuzde = float(input("M2 Makinesinin Kusurlu Kumaş Üretim Yüzdesini Giriniz:"))
M2KusurluKumasYuzde = M2KusurluKumasYuzde / 100
print("M2 Makinesinin Kusurlu Kumaş Üretim Yüzdesi: "+str(round(M2KusurluKumasYuzde,2))+"\n")

'''
M3 Makinesinin Kusurlu Kumaş Üretim Yüzdesinin Kullanıcdan Alınıp;
M3KusurluKumasYuzde Değişkenine Atanması
'''

M3KusurluKumasYuzde = float(input("M3 Makinesinin Kusurlu Kumaş Üretim Yüzdesini Giriniz:"))
M3KusurluKumasYuzde = M3KusurluKumasYuzde / 100
print("M3 Makinesinin Kusurlu Kumaş Üretim Yüzdesi: "+str(round(M3KusurluKumasYuzde,2))+"\n")

print("Üretilen Kumaş Kusurlu İse Kumaşın M1 Makinesinde Üretilme Olasılığı Nedir?\n\n")


time.sleep(3)

'''
1.Adım Üretilen Kumaşın Kusurlu Olma Olaslığı:
Bu Durum İstenilen Koşullu Olasılığın Örnek Uzay (Sample Space)\'ini Oluşturur.
'''

'''
Üretilen Kumaşın Kusurlu Olma Olasılığının Hesaplanıp;
KusurluKumas Değişkenine Atanması
'''

KusurluKumas = M1KumasYuzde*M1KusurluKumasYuzde + M2KumasYuzde*M2KusurluKumasYuzde + M3KumasYuzde*M3KusurluKumasYuzde
KusurluKumasMetin = "Üretilen Kumaşın Kusurlu Olma Olasılığı: {:.2}\n".format(KusurluKumas)
print(KusurluKumasMetin)

'''
2.Adım Üretilen Kumaşın M1 Makinesinde Üretilip Kusurlu Olma Olaslığı:
Bu Durum İstenilen Koşullu Olasılığın Olay Uzayı (Event Space)\'ini Oluşturur.
'''

'''
Üretilen Kumaşın M1 Makinesinde Üretilip Kusurlu Olma Olasılığının Hesaplanıp;
M1KusurluKumas Değişkenine Atanması
'''

M1KusurluKumas = M1KumasYuzde*M1KusurluKumasYuzde
M1KusurluKumasMetin = "Üretilen Kumaşın M1 Makinesinde Üretilip Kusurlu Olma Olasılığı: {:.4}\n".format(M1KusurluKumas)
print(M1KusurluKumasMetin)

'''
3. Adım Bir Olayın Olasılığı:
Olay Uzayı (Event Space)\'in Örnek Uzay (Sample Space)\'e Oranlanması Olduğu İçin
'''

'''
Üretilen Kumaşın Kusurlu Olduğu Bilndiğine Göre M1 Makinesinde Üretilmiş Olma Olasılığının Hesaplanıp;
KosulluKusurluKumas Değişkenine Atanması
'''

KosulluKusurluKumas = M1KusurluKumas / KusurluKumas
KosulluKusurluKumasMetin = "Üretilen Kumaşın Kusurlu Olduğu Bilndiğine Göre M1 Makinesinde Üretilmiş Olma Olasılığı: {:.4}".format(KosulluKusurluKumas)
print(KosulluKusurluKumasMetin)
```
